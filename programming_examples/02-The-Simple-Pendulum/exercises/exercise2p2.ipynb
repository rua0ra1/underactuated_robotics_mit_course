```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return -x**3 + 4*x**2 + 11*x - 30
```

```python
# Generate x values
x = np.linspace(-4, 6, 400)  # You can adjust the range and number of points
y=f(x)
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = -x^3 + 4x^2 + 11x - 30', color='blue')
plt.title('Plot of the Function y = -x^3 + 4x^2 + 11x - 30')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
def f(x,w):
    return (-x**3 + 4*x**2 + 11*x - 30)+w
```

```python
# Generate x values
x = np.linspace(-6, 6, 400)  # You can adjust the range and number of points

# Define different values of w to plot
w_values = [0,10,20,40,60]

```

```python
# Create a plot
plt.figure(figsize=(10, 8))
```

```python
# Plot the function for each value of w
for w in w_values:
    y = f(x, w)
    plt.plot(x, y, label=f'w = {w}')

# Add labels and title
plt.title('Plot of y = -x^3 + 4x^2 + 11x - 30 + w for Different Values of w')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
