```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x) = x + 2x + 3
def f(x,q0,q_dot_0):
    c=q0-(0.5*q_dot_0**2)
    q_dot=np.sqrt(2*(x-c))

    return q_dot
```

```python
# Generate x values for plotting
x_values = np.linspace(-100, 100, 4000)

# Calculate y values
y_values = f(x_values,0,0)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='u=1')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('two solutions of the system')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
